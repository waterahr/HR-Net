```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import csv

csv_file = csv.reader(open("./imagenet_models/GoogLeNet_RAP/binary51_b2(32)_lr0.0002_log.csv", "r"))
print(csv_file)
```

```python
loss = []
acc = []
val_acc = []
val_loss = []
csv_file = csv.reader(open("./imagenet_models/GoogLeNet_RAP/binary51_b2(32)_lr0.0002_log.csv", "r"))
for item in csv_file:
    if item[0] == 'epoch':
        continue
    print(item)
    acc.append(float(item[1]))
    loss.append(float(item[2]))
    val_acc.append(float(item[3]))
    val_loss.append(float(item[4]))
```

```python
fig = plt.figure(1)
fig.subplots_adjust(top=0.92, bottom=0.08, left=0.05, right=0.95, hspace=0.5, wspace=0.5)
ax_loss = fig.add_subplot(221)
ax_loss.set_xlabel("epoches")
ax_loss.set_ylabel("loss")
ax_loss.plot(loss, color='red', marker='x', linestyle='-', linewidth=1, markersize=0.1)

ax_acc = fig.add_subplot(222)
ax_acc.set_xlabel("epoches")
ax_acc.set_ylabel("accuracy")
ax_acc.set_ylim(0.75, 1)
ax_acc.plot(acc, color='red', marker='x', linestyle='-', linewidth=1, markersize=0.1)

ax_valloss = fig.add_subplot(223)
ax_valloss.set_xlabel("epoches")
ax_valloss.set_ylabel("val_loss")
ax_valloss.plot(val_loss, color='red', marker='x', linestyle='-', linewidth=1, markersize=0.1)

ax_valacc = fig.add_subplot(224)
ax_valacc.set_xlabel("epoches")
ax_valacc.set_ylabel("val_accuracy")
ax_valacc.set_ylim(0.80, 1)
ax_valacc.plot(val_acc, color='red', marker='x', linestyle='-', linewidth=1, markersize=0.1)

#labels = ('GoogLeNetSPP', 'hiarGoogLeNetSPP', 'GoogLeNet', 'hiarGoogLeNet', 'hiarBayesGoogLeNet', 'hiarGoogLeNetWAM')
#ax_valacc.legend(labels, loc='upper center', ncol=2, bbox_to_anchor= (-.3, -.5))#, borderaxespad=0, frameon=False, shadow=True
plt.show()
#fig.savefig("../results/"+model+"_log.png", dpi=256, bbox_inches="tight")
```
