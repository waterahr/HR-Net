# Calculate the ratio of the positive&negative sample

```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = ""

labels_list_file = r"/home/anhaoran/data/pedestrian_attributes_PETA/PETA/labels.txt" 
labels_list_data = open(labels_list_file)
lines = labels_list_data.readlines()
attributes_list = []
for line in lines:
    data = line.split()
    #print(attr)
    attributes_list.append(data[1])
print("The total attributes labeled: ", len(attributes_list))
print(attributes_list)
```

```python
import numpy as np
import pandas as pd

data = pd.read_csv(r"../results/PETA_labels_pd.csv")
len(data)
```

```python
data
```

```python
all_attributes = np.asarray(data)[:, 2:]
ratio_positive_negative_attributes = dict()
for i in range(all_attributes.shape[1]):
    tmp = all_attributes[:, i]
    pos = 0
    neg = 0
    for j in range(tmp.shape[0]):
        if tmp[j] == 1:
            pos += 1
        else:
            neg += 1
    ratio_positive_negative_attributes.update({attributes_list[i]: pos})
```

```python
print(ratio_positive_negative_attributes)
```

```python
import json

with open('../results/ratio_positive_samples_for_attributes.json','a') as outfile:
    json.dump(ratio_positive_negative_attributes, outfile, ensure_ascii=False)
    outfile.write('\n')
```

```python
f=open('../results/ratio_positive_samples_for_attributes.json',"r")
for line in f:
    data = json.loads(line)
print(data)
f.close()
```
