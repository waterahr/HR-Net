```python
import scipy.io as sio

!ls /home/anhaoran/data/pedestrian_attributes_PA-100K/annotation/
data = sio.loadmat('/home/anhaoran/data/pedestrian_attributes_PA-100K/annotation/annotation.mat')
data
```

```python
idx = 'attributes'
#print(data[idx])
print(type(data[idx]))
print(data[idx].shape)
attributes_list = []
for i in range(len(data[idx])):
    attributes_list.append(data[idx][i][0][0])
print(attributes_list)
```

```python
idx = 'train_label'
#print(data[idx])
print(type(data[idx]))
print(data[idx].shape)
train_label = data[idx]
print(train_label)
```

```python
idx = 'train_images_name'
#print(data[idx])
print(type(data[idx]))
print(data[idx].shape)
#print(data[idx])
train_images_name = []
for i in range(len(data[idx])):
    train_images_name.append("/home/anhaoran/data/pedestrian_attributes_PA-100K/data/release_data/release_data/"+str(data[idx][i][0][0]))
    if i <= 10:
        print(train_images_name[i])
#print(train_images_name)
```

```python
idx = 'val_label'
#print(data[idx])
print(type(data[idx]))
print(data[idx].shape)
val_label = data[idx]
print(val_label)

idx = 'val_images_name'
#print(data[idx])
print(type(data[idx]))
print(data[idx].shape)
#print(data[idx])
val_images_name = []
for i in range(len(data[idx])):
    val_images_name.append("/home/anhaoran/data/pedestrian_attributes_PA-100K/data/release_data/release_data/"+str(data[idx][i][0][0]))
    if i <= 10:
        print(val_images_name[i])
#print(train_images_name)
```

```python
idx = 'test_label'
#print(data[idx])
print(type(data[idx]))
print(data[idx].shape)
test_label = data[idx]
print(test_label)

idx = 'test_images_name'
#print(data[idx])
print(type(data[idx]))
print(data[idx].shape)
#print(data[idx])
test_images_name = []
for i in range(len(data[idx])):
    test_images_name.append("/home/anhaoran/data/pedestrian_attributes_PA-100K/data/release_data/release_data/"+str(data[idx][i][0][0]))
    if i <= 10:
        print(test_images_name[i])
#print(train_images_name)
```

```python
import numpy as np

all_images = np.vstack((np.asarray(train_images_name).reshape((-1,1)),
                        np.asarray(val_images_name).reshape((-1,1)),
                        np.asarray(test_images_name).reshape((-1,1))))
print(all_images.shape)
all_attributes = np.vstack((np.asarray(train_label),
                        np.asarray(val_label),
                        np.asarray(test_label)))
print(all_attributes.shape)
all_data = np.hstack((np.asarray(all_images).reshape((-1, 1)), np.asarray(all_attributes)))
print(all_data.shape)
```

```python
import pandas as pd

attributes_pd = pd.DataFrame(all_data)
attributes_pd
```

```python
attributes_pd.to_csv("../results/PA-100K_labels_pd.csv")
data_read = pd.read_csv("../results/PA-100K_labels_pd.csv")
data_read
```
