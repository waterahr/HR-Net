# picture size

```python
from keras.preprocessing import image

data_read = np.array(pd.read_csv(r"../results/RAP_labels_pd.csv"))[:, 1:]
length = len(data_read)
size = []
for i in range(length):
    img = image.load_img(data_read[i, 0])
    data_x = image.img_to_array(img)
    print(data_x.shape)
    size.append([data_x.shape[0], data_x.shape[1]])
```

```python
size = np.array(size)
print(size.shape)

print(np.mean(size[:, 0]))
print(np.median(size[:, 0]))

print(np.mean(size[:, 1]))
print(np.median(size[:, 1]))
```

```python
counts_height = np.bincount(size[:, 0])
#返回众数
print(np.max(counts_height))
print(np.argmax(counts_height))

counts_width = np.bincount(size[:, 1])
#返回众数
print(np.max(counts_width))
print(np.argmax(counts_width))
```
