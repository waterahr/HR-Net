```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = ""

attributes_list = ['is_male', 'has_long_hair', 'has_glasses', 'has_hat', 'has_t-shirt', 'has_long_sleeves', 'has_shorts', 'has_jeans', 'has_long_pants']
```

```python
from PIL import Image
import numpy as np
import math

images = []
attributes = []

labels_list_file = r"/home/anhaoran/data/pedestrian_attributes_berk/attributes_dataset/train/labels.txt" 
path = "/home/anhaoran/data/pedestrian_attributes_berk/attributes_dataset/train/"
labels_list_data = open(labels_list_file)
lines = labels_list_data.readlines()
for line in lines:
    #print(line)
    data = line.split()
    img = Image.open(path + data[0])
    #print(np.shape(img))
    if math.isnan(float(data[1])) or math.isnan(float(data[2])) or math.isnan(float(data[3])) or math.isnan(float(data[4])):
        print(line)
        continue
    img = img.crop((float(data[1]), float(data[2]), float(data[3]), float(data[4])))
    if(len(np.shape(img)) == 0):
        continue
    #print(data[0][:data[0].index(".jpg")])
    img_path = path+data[0][:data[0].index(".jpg")]+"_crop.jpg"
    images.append(img_path)
    img.save(img_path)
    label = np.zeros((9,))
    for i in range(5, len(data)):
        #print(data[i])
        if(float(data[i]) > 0):
            label[i-5] = 1
    attributes.append(label)
print(np.shape(attributes))
print(np.shape(images))
```

```python
all_data = np.hstack((np.asarray(images).reshape((-1, 1)), np.asarray(attributes)))
print(all_data)
print(all_data.shape)
```

```python
import pandas as pd

attributes_pd = pd.DataFrame(all_data)
attributes_pd
```

```python
attributes_pd.to_csv("../results/berk.csv")
```
