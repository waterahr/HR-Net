```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = ""

labels_list_file = r"/home/anhaoran/data/pedestrian_attributes_PETA/PETA/labels.txt" 
labels_list_data = open(labels_list_file)
lines = labels_list_data.readlines()
attributes_list = []
for line in lines:
    data = line.split()
    #print(attr)
    attributes_list.append(data[1])
print("The total attributes labeled: ", len(attributes_list))
print(attributes_list)
```

```python
import numpy as np

images = []
attributes = []
muticlass_attr = ['footwear', 'hair', 'lowerBody', 'upperBody']
colors = ['Black', 'Blue', 'Brown', 'Green', 'Grey', 'Orange', 'Pink', 'Purple', 'Red', 'White', 'Yellow']

#binary attribute, 0 - no and 1 - yes
#multi-class attribute, 1-11 for the color index and 0 for no exist
dataset_dir = "/home/anhaoran/data/pedestrian_attributes_PETA/PETA"
for dirpath, dirnames, filenames in os.walk(dataset_dir):
    #print(filenames)
    for file in filenames:
        if file == "Label.txt":
            #print(os.path.join(dirpath, file))
            f = open(os.path.join(dirpath, file))
            lines = f.readlines()
            #print(len(lines))
            for line in lines:
                data = line.split()
                images.append(dirpath + '/' + data[0])
                sample_label = np.zeros((65,))
                for i in range(1, len(data)):
                    attr = data[i]
                    idx = 1
                    if not attr in attributes_list:
                        print(attr)
                        if attr[:4] in muticlass_attr:
                            if attr[4:] in colors:
                                idx = colors.index(attr[4:]) + 1
                            else:
                                idx = 0
                            attr = attr[:4].lower()
                        elif attr[:8] in muticlass_attr:
                            if attr[8:] in colors:
                                idx = colors.index(attr[8:]) + 1
                            else:
                                idx = 0
                            attr = attr[:8].lower()
                        elif attr[:9] in muticlass_attr:
                            if attr[9:] in colors:
                                idx = colors.index(attr[9:]) + 1
                            else:
                                idx = 0
                            attr = attr[:9].lower()
                    if attr in attributes_list:
                        idx_a = attributes_list.index(attr)
                        sample_label[idx_a] = idx
                    else:
                        print(attr, " not found!!!")
                attributes.append(sample_label)

attributes = np.asarray(attributes)
print("The shape of attributes: ", attributes.shape)
print("The shape of images: ", np.shape(images))
```

```python
import os

muticlass_attr = ['footwear', 'hair', 'lowerBody', 'upperBody']
colors = ['Black', 'Blue', 'Brown', 'Green', 'Grey', 'Orange', 'Pink', 'Purple', 'Red', 'White', 'Yellow']

dataset_dir = "/home/anhaoran/data/pedestrian_attributes_PETA/PETA"
for dirpath, dirnames, filenames in os.walk(dataset_dir):
    #print(filenames)
    for file in filenames:
        if file == "Label.txt":
            #print(os.path.join(dirpath, file))
            f = open(os.path.join(dirpath, file))
            lines = f.readlines()
            #print(len(lines))
            for line in lines:
                data = line.split()
                for i in range(1, len(data)):
                    attr = data[i]
                    if not attr in attributes_list:
                        #print(attr)
                        if attr[:4] in muticlass_attr and attr[4:] in colors:
                            ;
                        elif attr[:8] in muticlass_attr and attr[8:] in colors:
                            ;
                        elif attr[:9] in muticlass_attr and attr[9:] in colors:
                            ;
                        else:
                            print(attr)
```

```python
all_images = []
all_attributes = []
for dirpath, dirnames, filenames in os.walk(dataset_dir):
    for file in filenames:
        if file != 'Label.txt' and file != 'labels.txt':
            all_images.append(os.path.join(dirpath, file))
            #print(file)
            #print(dirpath + '/' + file[:file.index('_')])
            if "CUHK" in dirpath:
                index = images.index(dirpath + '/' + file[:])
            else:
                index = images.index(dirpath + '/' + file[:file.index('_')])
            all_attributes.append(attributes[index])
```

```python
all_data = np.hstack((np.asarray(all_images).reshape((-1, 1)), np.asarray(all_attributes)))
print(all_data)
print(all_data.shape)
```

```python
import pandas as pd

attributes_pd = pd.DataFrame(all_data)
attributes_pd
```
