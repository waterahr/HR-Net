```python
import numpy as np
import pandas as pd
from keras.preprocessing import image

data = np.array(pd.read_csv(r"../results/PETA_labels_pd.csv"))[:, 1:]
length = len(data)
size = []
for i in range(length):
    img = image.load_img(data[i, 0])
    data_x = image.img_to_array(img)
    print(data_x.shape)
    size.append([data_x.shape[0], data_x.shape[1]])
```

```python
size = np.array(size)
```

```python
print(size.shape)
```

```python
print(np.mean(size[:, 0]))
print(np.median(size[:, 0]))
```
