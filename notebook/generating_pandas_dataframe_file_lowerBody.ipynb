```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = ""

labels_list_file = r"/home/anhaoran/data/pedestrian_attributes_PETA/PETA/labels.txt" 
labels_list_data = open(labels_list_file)
lines = labels_list_data.readlines()
attributes_list = []
for line in lines:
    data = line.split()
    #print(attr)
    attributes_list.append(data[1])
print("The total attributes labeled: ", len(attributes_list))
print(attributes_list)
```

```python
import numpy as np

index_in_pd = list(np.array(index) + 1)
index_in_pd
```

```python
new_data = []
all_data = np.array(all_data)[:, 1:]
all_data
```

```python
all_data.shape
```

```python
new_data = []
for i in range(len(all_data)):
    tmp = [all_data[i, 0]]
    #print(tmp)
    if np.sum(all_data[i, index_in_pd[1:6]]) == 1:
        tmp.append(0)
    elif np.sum(all_data[i, index_in_pd[1:6]]) != 0:
        print(i, '*', np.sum(all_data[i, index_in_pd[1:6]]))
    if np.sum(all_data[i, index_in_pd[6:8]]) == 1:
        tmp.append(1)
    elif np.sum(all_data[i, index_in_pd[6:8]]) != 0:
        print(i, '-', np.sum(all_data[i, index_in_pd[6:8]]))
    if len(tmp) != 2:
        print(i)
        continue
    if tmp[1] == 0:
        tmp.append(1)
    else:
        tmp[1] = 1
        tmp.append(0)
    for j in index_in_pd:
        tmp.append(all_data[i, j])
    new_data.append(tmp)
new_data = np.array(new_data)
new_data.shape
```
