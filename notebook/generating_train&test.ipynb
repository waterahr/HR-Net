```python
idxs = list(range(19000))
train_slice = random.sample(idxs, 9500)
for i in train_slice:
    idxs.remove(i)
val_slice = random.sample(idxs, 1900)
for i in val_slice:
    idxs.remove(i)
test_slice = idxs
print(len(train_slice), '\t', len(val_slice), '\t', len(test_slice))
```

```python
train = all_data_arr[train_slice]
val = all_data_arr[val_slice]
test = all_data_arr[test_slice]
print(train.shape)
```

```python
new_all_data_arr = np.vstack((train, val, test))
new_all_data_arr.shape
```

```python
attributes_pd = pd.DataFrame(new_all_data_arr[:, 1:])
attributes_pd
```
