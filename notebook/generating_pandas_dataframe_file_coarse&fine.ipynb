# Generating the coarse to fine attributes list

The former 7 for the coarse attributes and the next 65 for the fine attributes

```python
coarse_to_fine_attributes_list = {"coarse": ["accessory", "personal", "carrying", "hair", "footwear", "lowerbody", "upperbody"]}
coarse_to_fine_attributes_list.update({"fine": []})
print(coarse_to_fine_attributes_list)
```

```python
index_coarse = []
for i in range(len(coarse_to_fine_attributes_list['coarse'])):
    index_coarse.append([])
    coarse_to_fine_attributes_list["fine"].append([])
print(index_coarse)
```

```python
for attribute in attributes_list:
    if attribute[:4].lower() == "hair":
        idx = 0
    elif attribute[:8].lower() == "personal":
        idx = 1
    elif attribute[:8].lower() == "carrying":
        idx = 2
    elif attribute[:8].lower() == "footwear":
        idx = 3
    elif attribute[:9].lower() == "accessory":
        idx = 4
    elif attribute[:9].lower() == "lowerbody":
        idx = 5
    elif attribute[:9].lower() == "upperbody":
        idx = 6
    else:
        print(attribute)
    #index_coarse[idx].append(attributes_list.index(attribute))
    #coarse_to_fine_attributes_list["fine"][idx].append(attribute)
    #delete the multi-class attributes
    if attributes_list.index(attribute)<= 60:
        index_coarse[idx].append(attributes_list.index(attribute))
        coarse_to_fine_attributes_list["fine"][idx].append(attribute)
print(index_coarse)
print(coarse_to_fine_attributes_list)
```

```python
import numpy as np
all_data = np.array(all_data)[:, 1:]
print(all_data.shape)
```

```python
all_data_new = []
for i in range(len(all_data)):
    #print(all_data[i])
    tmp = [all_data[i][0]]
    for j in range(len(coarse_to_fine_attributes_list['coarse'])):
        if np.sum(all_data[i][(np.array(index_coarse[j]) + 1).tolist()]) != 0:
            tmp.append(1)
        else:
            tmp.append(0)
    for j in range(len(coarse_to_fine_attributes_list['coarse'])):
        for k in index_coarse[j]:
            tmp.append(all_data[i][k+1])
    all_data_new.append(tmp)
```

```python
np.shape(all_data_new)
```

```python
all_data_pd = pd.DataFrame(all_data_new)
all_data_pd.to_csv("../results/PETA_coarse_to_fine_labels_pd.csv")
```
